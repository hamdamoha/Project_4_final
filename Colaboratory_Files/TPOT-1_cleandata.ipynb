```python
 #Import our dependencies
import pandas as pd
import matplotlib as plt
import sklearn as skl
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tpot import TPOTClassifier
```

```python
#process the data from the csv file
#Import our input dataset
diab2_df = pd.read_csv('/content/sample_data/diabetes_upload.csv')
diab2_df.head()
```

```python
df_clean = diab2_df.copy()
```

```python
df_2 = df_clean[~(df_clean[df_clean.columns[1:5]]==0).any(axis=1)]
```

```python
#check zeros are removed
df_2
```

```python
#Remove Outcome target from features data
y = df_2.Outcome.values
X = df_2.drop(columns="Outcome").values
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 1502)
```

```python
model = TPOTClassifier(generations=100, 
                       population_size = 20,
                       verbosity = 2,
                       random_state = 40)

model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
```

```python
print(model.score(X_test,y_test))
```

```python
model.export('ml_best.py')
```
